```python
import numpy as np
from scipy import stats
import pandas as pd

data = pd.read_csv('weightheight.csv')
confidence = 0.95

column_name = 'Weight'  
data_column = data[column_name].dropna() 

# 1. Основные статистические показатели
n = len(data_column)
mean = np.mean(data_column)
se = stats.sem(data_column)

# 2. Вычисление доверительного интервала 
h = se * stats.t.ppf((1 + confidence) / 2., n - 1)
interval = (mean - h, mean + h)

print(f"95% ДИ: {interval}")
print(f"Среднее: {mean:.3f}")
print(f"Стандартная ошибка: {se:.3f}")
```
